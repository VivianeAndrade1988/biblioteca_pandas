```python
import pandas as pd
import seaborn as sns
```

```python
vendas= pd.read_excel("varejo.xlsx")
vendas.head()
```

```python
vendas.info()
```

```python
#qtd de compras por aplicativo
vendas.groupby("idcanalvenda").idcompra.nunique()
```

```python
#renomear APP para APlicativo
vendas["idcanalvenda"]=vendas ["idcanalvenda"].str.replace("APP","Aplicativo")
vendas.groupby("idcanalvenda").idcompra.nunique()
```

```python
# Qtd de compra por bandeiras
vendas.groupby("Nome_Departamento").idcompra.nunique()
```

```python
vendas.groupby("estado").idcompra.nunique()
```

```python
#colocando _ nas palavras com espaço
vendas["Nome_Departamento"]= vendas["Nome_Departamento"].str.replace(" ","_")
vendas.groupby("Nome_Departamento").idcompra.nunique()
```

```python
vendas.info()
```

```python
#mostrando dados nullos
vendas[vendas["estado"].isnull()]
```

```python
#deleta os nulos
vendas_sem_nulos=vendas.dropna()
vendas_sem_nulos.info()
```

```python
#SUbstituindo null por MG
vendas["estado"]= vendas ["estado"].fillna("MG")
vendas.info()
```

```python
#SUbstituir os que faltam preços
media_preco=vendas ["Preço"].mean()
vendas["Preço"].fillna(media_preco, inplace=True)
media_preco
```

```python
vendas.info()
```

```python
mediana_preco=vendas["Preço"].median()
mediana_preco
```

```python
#Substituir os nulls por 1000
vendas["Preço"].fillna(1000, inplace=True)
vendas.info()
```

```python
vendas.head()

```

```python
vendas[["Preço","Preço_com_frete"]].describe()
```

```python
#PREÇO NÃO PODE SER MAIOR QUE O PREÇO COM FRETE
preco_errado = vendas.query("Preço>Preço_com_frete")
preco_correto = vendas.query("Preço<Preço_com_frete")
```

```python
preco_correto.describe()
```

```python
preco_correto.head()
```

```python
preco_correto.query("Nome_Departamento=='Esporte_e_Lazer' and estado == 'SP'")
```

```python
preco_correto.query("Nome_Departamento=='Esporte_e_Lazer' and estado == 'SP'").describe()
```

```python
#DEPARTAMENTOS MAIS VENDIDOS
preco_correto.groupby("Nome_Departamento").idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
# media de preço com frete por nome de departamento
#casa decimais (round)
round(preco_correto.groupby("Nome_Departamento")["Preço_com_frete"].agg("mean").sort_values(ascending=False).reset_index(),2)
```

```python
#QTD DE VENDAS POR MÊS
preco_correto.groupby("Data").idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
#TRANSFORMAR DATA EM MÊS
preco_correto["mês"] = preco_correto["Data"].dt.month
preco_correto.head()
```

```python

preco_correto.groupby("mês").idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
cliente= pd.read_excel("cliente_varejo.xlsx")
cliente.head()
```

```python
preco_correto.head()
```

```python
cliente[["idade","renda"]].describe()
```

```python
#transformando renda em float
cliente= cliente.astype({"renda":"float"})
cliente.info()
```

```python
#join, união de tabelas
#merge= união
vendas_cliente= preco_correto.merge(cliente,how= "left", on="cliente_Log")
vendas_cliente.head()
```

```python
#MÉDIA DE RENDA PARA CADA TIPO DE CANAL DE VENDA

agg_idcanal_renda= round(vendas_cliente.groupby("idcanalvenda")["renda"].agg("mean").sort_values(ascending=False).reset_index(),2)
agg_idcanal_renda.head()
```

```python
#MÉDIA DE IDADE DE CLIENTES POR BANDEIRA
agg_idade_bandeira= round(vendas_cliente.groupby("bandeira")["idade"].agg("mean").sort_values(ascending=False).reset_index(),2)
agg_idade_bandeira.head()
```

```python
import matplotlib.pyplot as plt
```

```python
fig=plt.figure(figsize=(7,4))
plt.bar(agg_idade_bandeira["bandeira"],agg_idade_bandeira["idade"],color="black")
#titulo Y
plt.ylabel("Média de idade")
plt.title("Idade média por bandeira",loc="center",fontsize=14,fontweight="bold",color="black");
```

```python
fig=plt.figure(figsize=(7,4))
plt.bar(agg_idcanal_renda["idcanalvenda"],agg_idcanal_renda["renda"],color="blue")
#titulo Y
plt.ylabel("Média de renda ")
plt.title("Renda média por canal de venda",loc="center",fontsize=14,fontweight="bold",color="blue");
```

```python
venda_por_data= preco_correto.groupby("Data").idcompra.nunique().reset_index()
venda_por_data
```

```python
fig=plt.figure(figsize=(20,8))

plt.xlabel("Data da Venda")
plt.ylabel("Qtd de vendas")
x= venda_por_data["Data"]
y=venda_por_data["idcompra"]

plt.plot(x,y,color="blue", linewidth="2")
plt.show
```

```python
agg_depto_preco=round(preco_correto.groupby("Nome_Departamento")["Preço_com_frete"].agg("mean").sort_values(ascending=False).reset_index(),2)
agg_depto_preco.head()
```

```python
fig=plt.figure(figsize=(25,8))
plt.bar(agg_depto_preco["Nome_Departamento"],agg_depto_preco["Preço_com_frete"],color="blue")
#titulo Y
plt.ylabel("Preço médio ")
plt.title("Preço médio por departamento",loc="center",fontsize=14,fontweight="bold",color="blue");
```
